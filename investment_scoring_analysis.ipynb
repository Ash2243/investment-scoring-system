```python
import pandas as pd
import os
```

```python
path = r"C:\Users\VAIBHAV\Downloads\archive (5)"

all_data = []

for file in os.listdir(path):
    if file.endswith(".csv"):
        if file == 'stock_metadata.csv':
            continue
        file_path = os.path.join(path, file)

        df = pd.read_csv(file_path)
        
        symbol = file.replace(".csv", "")
        df['Firm'] = symbol
        
        all_data.append(df)
```

```python
df = pd.concat(all_data, ignore_index=True)
df.head(50)
```

```python
df.info()
```

```python
df['Trades']=df['Trades'].fillna(0)
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'], 
      dayfirst = True     
)
df['year']=df['Date'].dt.year
df.head()
```

```python
col = df.pop('Firm')
df.insert(2, 'Firm', col)
df.head()
```

```python
col = df.pop('year')
df.insert(1, 'year', col)
df.head()
```

```python
df['Volume'] = pd.to_numeric(df['Volume'].astype(str).str.replace(',', ''), errors='coerce')
df.info()
```

```python
import numpy as np
```

```python
df['daily_return'] = (df['Close']-df['Prev Close'])/df['Prev Close']
df['daily_return'] = df['daily_return'].replace(np.inf, np.nan)
df.head()
```

```python
df[df['Prev Close']==0]
```

```python
df['volatility']=df.groupby('Firm')['daily_return'].transform('std')
df.head()
```

```python
df=df.sort_values(by=['Firm','Date'])
```

```python
df['MA_20'] = df.groupby('Firm')['daily_return'].transform(lambda x: x.rolling(20).mean())
df['MA_50'] = df.groupby('Firm')['daily_return'].transform(lambda x: x.rolling(50).mean())
df.info()
```

```python
df['trend'] = np.where(
    df['Close'] > df['MA_50'], 'Upward',
    np.where(df['Close'] < df['MA_50'], 'Downward', 'Neutral')
)
```

```python
df.tail()
```

```python
df_new=df.groupby('Firm').agg({'daily_return' : ['mean'], 'volatility' : ['mean'],'trend': lambda x: (x == 'Upward').mean()})
df_new.columns
```

```python
df_new.sort_values(by=[('daily_return',     'mean')],ascending=False)
```

```python
df_new.rename(columns={
    "('daily_return',     'mean')": "avg_return",
    "(  'volatility',     'mean')": "avg_volatility",
    "(       'trend', '<lambda>')":"%upward_trend"}, inplace=True)
df_new
```

```python
df_new.columns = ['_'.join(col) for col in df_new.columns]
```

```python
df_new.columns
```

```python
df_new=df_new.drop(columns='avg_return_')
```

```python
df_new.sort_values(by='avg_return',ascending=False)
```

```python
df_new['return_score'] = ((df_new['avg_return'] - df_new['avg_return'].min()) / (df_new['avg_return'].max() - df_new['avg_return'].min()))*0.5
df_new['volatility_score'] =(1-((df_new['avg_volatility'] - df_new['avg_volatility'].min()) / (df_new['avg_volatility'].max() - df_new['avg_volatility'].min())))*0.3
df_new['trend_score'] = ((df_new['%upward_trend'] - df_new['%upward_trend'].min()) / (df_new['%upward_trend'].max() - df_new['%upward_trend'].min()))*0.2
df_new['investment_score']=df_new['return_score']+df_new['volatility_score']+df_new['trend_score']
df_new
```

```python
df_new=df_new.reset_index(col_fill='Firm')
```

```python
df_new.to_excel('Metadata_new.xlsx',index=False)
df_new
```
